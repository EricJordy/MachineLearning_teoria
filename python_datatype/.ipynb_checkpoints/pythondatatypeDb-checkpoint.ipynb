## 2. Literales, variables y tipos de datos básicos

De forma muy genérica, al ejecutarse un programa Python, simplemente se realizan operaciones sobre objetos.

Estos dos términos son fundamentales:

- **Objetos:** cualquier tipo de datos (números, caracteres o datos más complejos)
- **Operaciones:** cómo manipulamos estos datos.

### Ejemplo

```python
4 + 3
```

### 2.1 Literales

- Python tiene una serie de tipos de datos integrados en el propio lenguaje.
- Los literales son expresiones que generan objetos de estos tipos.
- Estos objetos pueden ser simples o compuestos.
- También pueden ser: <br>
  - Simples o compuestos <br>
  - Mutables o inmutables.

### Literales simples

- Enteros.
- Decimales o punto flotante.
- Booleanos.

```python
print(4)              # número entero
print(4.2)            # número en coma flotante
print('Hello world!') # string
print(False)          # booleano
```

### Literales compuestos

- Tuplas
- Listas
- Diccionarios
- Conjuntos

```python
print([1, 2, 3, 3])                         # lista - mutable
print({'Nombre': 'John Doe', 'edad': 30})   # diccionario - mutable
print({1, 2, 3, 3})                         # conjunto - mutable
print((4, 5))                               # tupla - inmutable
2, 4                                        # tupla
```

### 2.2 Variables

- Referencias a objetos.
- Las variables y los objetos se almacenan en diferentes zonas de memoria.
- Las variables siempre referencian a objetos y nunca a otras variables.
- Objetos sí que pueden referenciar a otros objetos. Ejemplo: listas.
- Sentencia de asignación: <br>

`<nombre_variable> = <objeto>`

```python
# # Asignación de variables
a = 5
print(a)
```

```python
a = 1               # entero
b = 4.0             # coma flotante
c = "ITQ"           # string
d = 10 + 1j         # numero complejo
e = True #False     # boolean
f = None            # None

# visualizar valor de las variables y su tipo
print(a)
print(type(a))

print(b)
print(type(b))

print(c)
print(type(c))

print(d)
print(type(d))

print(e)
print(type(e))

print(f)
print(type(f))
```

- Las variables no tienen tipo.
- Las variables apuntana objetos que sí lo tienen .
- Dado que Python es un lenguaje de tipado dinámico, la misma variable puede apuntar, en
momentos diferentes de la ejecución del programa, objetos de diferente tipo.

```python
a = 3
print(a)       # Muestra el valor
print(type(a)) # Muestra el tipo de dato

a = 'Pablo García'
print(a)       # Muestra el valor
print(type(a)) # Muestra el tipo de dato

a = 4.5
print(a)       # Muestra el valor
print(type(a)) # Muestra el tipo de dato
```

- Garbagecollection: Cuando un objeto de jade estar referenciado, se elimina automáticamente. 
#### Identificadores
- Podemos obtener unidentificador únicoparalos objetos referenciados por variables.
- Este identificador se obtiene apartir de la dirección de memoria.

```python
a = 3
print(id(a)) # Muestra la dirección en memoria

a = 'Pablo García'
print(id(a)) # Muestra la dirección en memoria

a = 4.5
print(id(a)) # Muestra la dirección en memoria
```

- Las variables pueden aparecer en expresiones.

```python
a = 3
b = 5
print (a+ b) # Suma los dos valores
```

```python
c = a + b
print(c) # Muestra el resultado de la suma
print(id(c)) # Muestra la dirección en memoria
```

### 2.3 Tipos de datos básicos
### Bool
- 2posibles valores: ‘True’ o ‘False’

```python
a = False
b = True

print(a)       # Muestra el valor booleano
print(type(a)) # Muestra el tipo de dato

print(b)       # Muestra el valor booleano
print(type(b)) # Muestra el tipo de dato
```

- ‘True’ y ‘False’ también son objetos que se guardan en caché, al igual que los enteros pequeños.

```python
a = True
b = True

print(id(a)) # Muestra la dirección en memoria
print(id(b)) # Muestra la dirección en memoria

print(a is b) # Compara si son el mismo objeto
print(a == b) # Compara si tienen el mismo valor
```

### Números

```python
print(2)      # Enteros, sin parte fraccional.
print(3.4)    # Números en coma flotante, con parte fraccional.
print(2+4j)   # Números complejos.
print(1/2)    # Numeros racionales.
```

### Strings
- Cadenas de caracteres.
- Son secuencias: la posición de los caracteres es importante.
- Son immutables: las operaciones sobre strings no cambian el string original

```python
s = 'John "ee" Doe'
print(s[0])     # Primer carácter del string.
print(s[-1])    # Último carácter del string.
print(s[1:8:2]) # Substring desde el segundo carácter (inclusive) hasta el␣ octavo (exclusive). Esta técnica se la conoce como 'slicing'.
print(s[:])     # Todo el string.
print(s + "e")  # Concatenación.
```

### 2.4 Conversión entre tipos
- Aveces queremos que un objeto sea de un tipo específico.
- Podemos obtener objetos de un tipo a partir de objetos de un tipo diferente (casting).

```python
a = int(2.8)     # a será 2
b = int("3")     # b será 3
c = float(1)     # c será 1.0
d = float("3")   # d será 3.0
e = str(2)       # e será '2'
f = str(3.0)     # f será '3.0'
g = bool("a")    # g será True
h = bool("")     # h será False
i = bool(3)      # i será True
j = bool(0)      # j será False
k = bool(None)
print(a)
print(type(a))
print(b)
print(type(b))
print(c)
print(type(c))
print(d)
print(type(d))
print(e)
print(type(e))
print(f)
print(type(f))
print(g)
print(type(g))
print(h)
print(type(h))
print(i)
print(type(i))
print(j)
print(type(j))
print(k)
```

```python
print(7/4) #Divisiónconvencional.Resultadodetipo'float'
print(7//4) #Divisiónentera.Resultadodetipo'int'
print(int(7/4)) #Divisónconvencional.Conversióndelresultadode'float'a␣'int'
```

```python
x = 0b01100110
y = 0b00110011

print("Not x = " + bin(~x))      # Aplica el complemento a nivel de bits
print("x and y = " + bin(x & y)) # Realiza AND a nivel de bits
print("x or y = " + bin(x | y))  # Realiza OR a nivel de bits
print("x xor y = " + bin(x ^ y)) # Realiza XOR a nivel de bits
print("x << 2 = " + bin(x << 2)) # Desplaza los bits dos posiciones a la izquierda
print("x >> 2 = " + bin(x >> 2)) # Desplaza los bits dos posiciones a la derecha
```

## 3 Tipos de datos compuestos (colecciones)
### 3.1 Listas
- Una colección de objetos.
- Mutables.
- Tipos arbitrarios heterogeneos.
- Puede contener duplicados.
- Notienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.
- Los elementos van ordenados por posición.
- Se acceden usando la sintaxis: [index].
- Los índices van de 0 a n-1, donde n es el número de elementos de la lista.
- Son un tipo de Secuencia, al igual que los strings; por lo tanto, el orden (es decir, la posición
de los objetos de la lista) es importante.
- Soportan anidamiento.
- Son una implementación del tipo abstracto de datos: Array Dinámico.

<img src="img/data_types/list.png" width="650">


### Operaciones con listas
- Creación de listas.

```python
letras = ['a', 'b', 'c', 'd']     # Crea una lista de letras
palabras = 'Hola mundo como estas'.split()    # Divide el texto en una lista
numeros = list(range(7))    # Crea una lista de números

print(letras)         # Muestra la lista de letras
print(palabras)      # Muestra la lista de palabras
print(numeros)       # Muestra la lista de números
print(type(numeros)) # Muestra el tipo de dato
```

```python
# # Pueden contener elementos arbitrarios / heterogeneos
mezcla = [1, 3.4, 'a', None, False]
print(mezcla)
print(len(mezcla)) # len me da el tamaño de la lista número de elementos
```

```python
# # Pueden incluso contener objetos más "complejos"
lista_con_funcion = [1, 2, len, pow]
print(lista_con_funcion)
```

```python
# # Pueden contener duplicados
lista_con_duplicados = [1, 2, 3, 3, 3, 4]
print(lista_con_duplicados)
```

- Obtención de la longitud de una lista

```python
letras = ['a', 'b', 'c', 'd', 1]
print(len(letras))
```

- Acceso a un elemento de una lista

```python
letras = ['a', 'b', 'c', 'd', 1]
print(letras[2]) # Muestra el tercer elemento
print(letras[-5]) # Genera error por índice fuera de rango
print(letras[0]) # Muestra el primer elemento
```

- Slicing: obtención de un fragmento de una lista, devuelve una copia de una parte de la lista
    - Sintaxis: lista [ inicio : fin : paso ]

```python
letras = ['a', 'b', 'c', 'd', 'e']

print(letras[1:3])   # Muestra elementos del índice 1 al 2
print(letras[:3])    # Muestra los tres primeros elementos
print(letras[:-1])   # Muestra todos excepto el último
print(letras[2:])    # Muestra desde el índice 2 hasta el final
print(letras[:])     # Muestra toda la lista
print(letras[::2])   # Muestra elementos de dos en dos
```

```python
letras = ['a', 'b', 'c', 'd']

print(letras)      # Muestra la lista
print(id(letras))  # Muestra la dirección en memoria

a = letras[:]      # Crea una copia de la lista
print(a)           # Muestra la copia
print(id(a))       # Muestra la dirección en memoria

print(letras.copy())   # Crea y muestra otra copia
print(id(letras.copy()))   # Muestra la dirección en memoria de la copia
```

- Añadir un elemento al final de la lista

```python
letras.append('e') # Agrega un elemento al final
print(letras)      # Muestra la lista
print(id(letras))  # Muestra la dirección en memoria
```

```python
letras += 'e'   # Agrega el elemento usando el operador +=
print(letras)   # Muestra la lista
print(id(letras)) # Muestra la dirección en memoria
```

- Insertar en posición

```python
print(len(letras))    # Muestra la cantidad de elementos
letras.insert(1, 'g') # Inserta un elemento en la posición indicada
print(len(letras)) # Muestra la nueva cantidad de elementos
print(letras)      # Muestra la lista
print(id(letras))  # Muestra la dirección en memoria
```

```python
# Ojo con la diferencia entre modificación individual y múltiple. Asignación individual de lista crea anidamiento.

numeros = [1, 2, 3]        # Crea una lista
numeros[1] = [10, 20, 30]  # Reemplaza un elemento por una lista
print(numeros)             # Muestra la lista
print(numeros[1][0])       # Muestra el primer elemento de la lista anidada

numeros[1][2] = [100, 200] # Reemplaza un elemento por otra lista
print(numeros)             # Muestra la lista
print(numeros[1][2][1])    # Muestra el segundo elemento de la lista anidada
```

- Encontrar índice de un elemento.

```python
letras = ['a', 'b', 'c', 'c'] # Crea una lista

if 'a' in letras:    # Comprueba si el elemento existe
    print(letras.index('a'))   # Muestra el índice del elemento

print(letras)       # Muestra la lista
```

- Concatenar listas.

```python
lacteos = ['queso', 'leche']     # Crea una lista de lácteos
frutas = ['naranja', 'manzana']  # Crea una lista de frutas

print(id(lacteos))    # Muestra la dirección en memoria
print(id(frutas))     # Muestra la dirección en memoria

compra = lacteos + frutas   # Une las dos listas
print(id(compra))           # Muestra la dirección en memoria
print(compra)               # Muestra la lista resultante
```

```python
# Concatenación sin crear una nueva lista

frutas = ['naranja', 'manzana']   # Crea una lista de frutas
print(id(frutas))    # Muestra la dirección en memoria

frutas.extend(['pera', 'uvas'])   # Agrega varios elementos al final
print(frutas)        # Muestra la lista
print(id(frutas))    # Muestra la dirección en memoria
```

```python
# Anidar sin crear una nueva lista

frutas = ['naranja', 'manzana']   # Crea una lista de frutas

print(id(frutas))    # Muestra la dirección en memoria

frutas.append(['pera', 'uvas'])   # Agrega una lista como un solo elemento
print(frutas)                     # Muestra la lista
print(id(frutas))                 # Muestra la dirección en memoria
```

### Alias/Referencias en las listas
- Son mutables y las asignaciones a un objeto alteran el primero 
- Pasar listas a funciones puede suponer un riesgo
- Clonar o copiar listas

```python
lista = [2,4,16,32] # Crea una lista de números

ref = lista         # Asigna la referencia de la lista

print(id(lista))    # Muestra la dirección en memoria
print(id(ref))      # Muestra la dirección en memoria

ref[2] = 64         # Modifica un elemento de la lista

print(ref)          # Muestra la lista modificada
print(lista)        # Muestra la lista original modificada
```

```python
lista = [2000, 4, 16, 32] # Crea una lista de números

copia = lista[:]          # Crea una copia de la lista
copia = lista.copy()      # Crea otra copia de la lista

copia[2] = 64             # Modifica un elemento de la copia

print(lista)              # Muestra la lista original
print(copia)              # Muestra la copia modificada

print(id(lista))          # Muestra la dirección en memoria de la lista
print(id(copia))          # Muestra la dirección en memoria de la copia

print(id(lista[2]))       # Muestra la dirección en memoria del elemento original
print(id(copia[2]))       # Muestra la dirección en memoria del elemento modificado
```

```python
# Listas anidadas se copian por referencia

lista = [2, 4, 16, 32, [34, 10, [5,5]]] # Crea una lista con elementos anidados
copia = lista.copy() # Crea una copia superficial de la lista

copia[3] = 20      # Modifica un elemento de la copia
copia[4][0] = 28   # Modifica un elemento de la lista anidada

print(copia) # Muestra la copia
print(lista) # Muestra la lista original
```

```python
lista = [0, 1, [10, 20]] # Crea una lista con una lista anidada
print(id(lista))    # Muestra la dirección en memoria

lista2 = [10, 20]   # Crea una lista
lista = [0, 1, lista2] # Asigna la lista anidada

copia = lista[:]    # Crea una copia superficial
print(copia)        # Muestra la copia
print(id(copia))    # Muestra la dirección en memoria

print(id(lista[2])) # Muestra la dirección en memoria de la lista anidada
print(id(copia[2])) # Muestra la dirección en memoria de la lista anidada copiada

copia[2][0] = 40 # Modifica un elemento de la lista anidada
print(copia)     # Muestra la copia modificada
print(lista)     # Muestra la lista original
```

```python
# Evitar que listas anidadas se copien por referencia

import copy # Importa el módulo para copias profundas

lista = [2, 4, 16, 32, [34, 10, [5,5]]] # Crea una lista con elementos anidados

copia = copy.deepcopy(lista)            # Crea una copia profunda de la lista

copia[0] = 454        # Modifica un elemento de la copia
copia[4][2][0] = 64   # Modifica un elemento anidado de la copia

print(lista)    # Muestra la lista original
print(copia)    # Muestra la copia modificada

print(f"{id(lista)}- {id(copia)}")         # Muestra la dirección en memoria de ambas listas
print(f"{id(lista[4])}- {id(copia[4])}")   # Muestra la dirección en memoria de las listas anidadas
```

### 3.2 Diccionarios
- Colección de parejas clave-valor.
- Son mutables.
- Claves:
    - Cualquier objeto inmutable.
    - Sólo pueden aparecer una vez en el diccionario.
- Valores:
    - Sin restricciones. Cualquier objeto (enteros, strings, listas, etc.) puede hacer de valor.
- Desde la versión 3.7 están ordenados.
- Se pueden ver como listas indexadas por cualquier objeto inmutable, no necesariamente por números enteros.
- Adiferencia de las listas, no son secuencias. Son mappings

### Operaciones con diccionarios
- Creación de diccionarios.

```python
# Creación simple, usando una expresión literal.
persona = {'DNI' : '11111111D', 'Nombre' : 'Carlos', 'Edad' : 34}
print(persona)
```

```python
# Creación uniendo dos colecciones.

nombres = ['Pablo', 'Manolo', 'Pepe', 'Juan'] # Crea una lista de nombres
edades = [52, 14, 65]      # Crea una lista de edades

datos = dict(zip(nombres, edades))  # Une ambas listas en un diccionario
print(datos)               # Muestra el diccionario
```

```python
# Creación pasando claves y valores a la función 'dict'
persona2 = dict(nombre='Rosa', apellido='Garcia')
print(persona2)
```

```python
# Creación usando una lista de tuplas de dos elementos.
persona2 = dict([('nombre', 'Rosa'), ('apellido','Garcia')])
print(persona2)
```

```python
# Creación incremental por medio de asignación (como las claves no existen, se crean nuevos items)

persona = {}        # Crea un diccionario vacío

persona['DNI'] = '11111111D' # Agrega la clave DNI
persona['Nombre'] = 'Carlos' # Agrega la clave Nombre
persona['Edad'] = 34         # Agrega la clave Edad

print(persona['DNI'])        # Muestra el valor de la clave DNI

persona['DNI'] = '222222222' # Actualiza el valor de la clave DNI

print(persona)               # Muestra el diccionario
```

- Acceso a un valor a través de la clave

```python
persona = {'DNI' : '11111111D', 'Nombre' : 'Eric', 'Edad' : 20}
print(persona['Nombre'])
```

```python
# Acceso a claves inexistentes o por índice produce error

# persona[1]
print(persona['Nombre']) # Muestra el valor de la clave Nombre

if 'Trabajo' in persona: # Comprueba si existe la clave Trabajo
    print(persona['Trabajo']) # Muestra el valor de la clave
else:
    print("En el paro") # Muestra un mensaje si no existe

persona.get('Trabajo', "En el paro") # Obtiene el valor o devuelve el valor por defecto
```

- Modificación de un valor a través de la clave

```python
persona = {'DNI' : '11111111D', 'Nombre' : 'Carlos', 'Edad' : 34} # Crea un diccionario

persona['Nombre'] = 'Fernando' # Actualiza el nombre
persona['Edad'] += 1           # Incrementa la edad en 1

print(persona)                 # Muestra el diccionario
```

- Añadir un valor a través de la clave

```python
persona = {'DNI' : '11111111D', 'Nombre' : 'Carlos', 'Edad' : 34}
persona['Ciudad'] = 'Valencia'

print(persona)
```

- Eliminación de un valor a través de la clave

```python
del persona['Ciudad']
value = persona.pop('Edad')

print(persona)
print(value)
```

- Comprobación de existencia de clave

```python
print('Nombre' in persona)
print('Apellido' in persona)
```

- Recuperación del valor de una clave, indicando valor por defecto en caso de ausencia

```python
persona = {'Nombre': 'Carlos'} # Crea un diccionario

value = persona.get('Nombre') # Obtiene el valor de la clave Nombre
print(value)                  # Muestra el valor obtenido

value = persona.get('Estatura', 180) # Obtiene el valor o devuelve 180
print(value)                  # Muestra el valor obtenido
```

- Anidamiento

```python
persona = {
'Trabajos' : ['desarrollador', 'gestor'],
'Direccion' : {'Calle' : 'Pintor Sorolla', 'Ciudad' : 'Valencia'}
}

print(persona['Direccion'])
print(persona['Direccion']['Ciudad'])
```

- Métodos items, keys y values

```python
persona = {'DNI' : '11111111D', 'Nombre' : 'Carlos', 'Edad' : 34}

print(list(persona.items()))
print(list(persona.keys()))
print(list(persona.values()))
```

```python
dict_simple = {'ID' : 'XCSDe1194', 'Nombre' : 'Carlos', 'Edad' : 34}

# iterar sobre las claves
print('Claves\n')
for key in dict_simple.keys(): # o dict_simple
  print(key)

print('\nValores\n')
# iterar sobre los valores
for value in dict_simple.values():
   print(value)
```

```python
dict_simple = {
    'ID': 'XCSDe1194',
    'Nombre': 'Carlos',
    'Edad': 34
}

# Iterar sobre clave y valor con items()
for key, value in dict_simple.items():
    print('Clave: ' + str(key) + ', valor: ' + str(value))

# items() devuelve tuplas
for item in dict_simple.items():
    print('Item: ' + str(item))

# Usar zip para iterar sobre claves y valores
for key, value in zip(dict_simple.keys(), dict_simple.values()):
    print('Clave: ' + str(key) + ', valor: ' + str(value))
```

### 3.3 Sets (Conjuntos)
Al igual que las listas:

- Colección de elementos.
- Tipos arbitrarios.
- Mutables.
- Notienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.

A diferencia de las listas:
- Nopuede tener duplicados.
- Se definen por medio de llaves.
- Los elementos no van ordenados por posición. No hay orden establecido.
- Solo pueden contener objetos inmutables.
- Nosoportan anidamiento.
### Operaciones con conjuntos
- Creación de conjuntos.

```python
set1 = {0, 1, 1, 2, 3, 4, 4} # Crea un conjunto eliminando duplicados
print(set1)             # Muestra el conjunto

set2 = {'user1', 12, 2} # Crea un conjunto con distintos tipos de datos
print(set2)             # Muestra el conjunto

set3 = set(range(7)) # Crea un conjunto con valores del 0 al 6
print(set3)          # Muestra el conjunto

set4 = set([0, 1, 2, 3, 4, 0, 1]) # Convierte una lista en un conjunto
print(set4)                       # Muestra el conjunto
```

```python
# Observa la diferencia entre listas y conjuntos
s = 'aabbc'
print(list(s))
print(set(s))
```

- Acceso por índice genera error.

```python
set1 = {0, 1, 2}

lista = list(set1)
print(lista[0])
```

- Unión, intersección y diferencia

```python
set1 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21} # Crea un conjunto
set2 = set([0, 1, 2, 3, 4, 42]) # Crea un conjunto desde una lista

# Unión
print(set1 | set2) # Muestra la unión de los conjuntos

# Intersección
print(set1 & set2) # Muestra los elementos comunes

# Diferencia
print(set1 - set2) # Muestra los elementos de set1 que no están en set2
print(set2 - set1) # Muestra los elementos de set2 que no están en set1
```

```python
a = [1,2]
b = [2,3]
print(set(a) & set(b))
```

```python
# Además de los operadores, que operan únicamente con Sets, también se pueden␣usar métodos que pueden operar sobre cualquier objeto iterable.
conjunto = {0, 1, 2}
lista = [1, 3, 3]

print(conjunto.union(lista))
print(conjunto.intersection(lista))
print(conjunto.difference(lista))
```

- comparación de conjuntos

```python
set1 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21}
set2 = set([0, 1, 2, 3, 4])

print(set2.issubset(set1))
print(set1.issuperset(set2))
print(set1.isdisjoint(set2))
```

- Pertenencia

```python
words = {'calm', 'balm'}
print('calm' in words)
```

- Anidamiento

```python
# Los conjuntos no soportan anidamiento, pero como permite elementos␣inmutables, se pueden "anidar" tuplas.

nested_set = {1, (1, 1, 1), 2, 3, (1,1,1)}
print(nested_set)
```

- Modificación de conjuntos

```python
# A través de operador de asignación

set1 = {'a', 'b', 'c'} # Crea un conjunto
set2 = {'a', 'd'} # Crea otro conjunto

set1 |= set2 # Realiza la unión y actualiza el conjunto
set1 &= set2 # Realiza la intersección y actualiza el conjunto
set1 -= set2 # Realiza la diferencia y actualiza el conjunto

print(set1) # Muestra el conjunto resultante
```

```python
# A través del método update()

set1 = {'a', 'b', 'c'} # Crea un conjunto
set2 = {'a', 'd'} # Crea otro conjunto

set1.update(set2) # Realiza la unión del conjunto
set1.intersection_update(set2) # Conserva los elementos comunes
set1.difference_update(set2) # Elimina los elementos de set2

print(set1) # Muestra el conjunto resultante
```

```python
# A través de métodos 'add' y 'remove'.

set1 = {'a', 'b', 'c'}

set1.add('d')
set1.remove('a')

print(set1)
```

### 3.4 Tuplas
Al igual que las listas:
- Colección de elementos.
- Tipos arbitrarios.
- Puede contener duplicados.
- Notienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.
- Los elementos van ordenados por posición.
- Acceso a través de la sintaxis: [index]
- Índices van de 0 a n-1, donde n es el número de elementos de la tupla.
- Son secuencias donde el orden de los elementos importa.
- Soportan anidamiento.

A diferencia de las listas:
- Se definen por medio de paréntesis.
- Inmutables.
### ¿Por qué tuplas?
- Representación de una colección fija de elementos (por ejemplo, una fecha).
- Pueden usarse en contextos que requieren inmutabilidad (por ejemplo, como claves de un
diccionario).
### Operaciones con tuplas
- Creación de tuplas. 

```python
tuple1 = ('Foo', 34, 5.0, 34) # Crea una tupla
print(tuple1) # Muestra la tupla

tuple2 = 1, 2, 3 # Crea una tupla sin paréntesis
print(tuple2) # Muestra la tupla

tuple3 = tuple(range(10)) # Crea una tupla con valores del 0 al 9
print(tuple3) # Muestra la tupla

tuple4 = tuple([0, 1, 2, 3, 4]) # Convierte una lista en tupla
print(tuple4) # Muestra la tupla
```

```python
# Ojo con las tuplas de un elemento. Los paréntesis se intepretan como␣indicadores de precedencia de operadores.
singleton_number = (1)
type(singleton_number)
```

```python
# Creación de tupla de un elemento.
singleton_tuple = (1,)
print(type(singleton_tuple))
print(singleton_tuple)
```

- Obtención del número de elementos

```python
print(len(tuple1))
```

- Acceso por índice

```python
tuple1 = ('Foo', 1, 2, 3)

print(tuple1[0]) # Primer elemento
print(tuple1[len(tuple1)-1]) # Último elemento

print(tuple1[-1]) # Índices negativos comienzan desde el final
print(tuple1[-len(tuple1)]) # primer elemento
```

- Asignación a tuplas falla. Son immutables

```python
print(tuple1[0])   #Correcto
```

```python
print(id(tuple1)) # Muestra la dirección en memoria

lista = list(tuple1)  # Convierte la tupla en lista
lista[0] = 'bar'      # Modifica el primer elemento
tuple1 = tuple(lista) # Convierte la lista nuevamente en tupla

print(id(tuple1)) # Muestra la nueva dirección en memoria
print(tuple1)     # Muestra la tupla modificada
```

- Contar número de ocurrencias de un elemento

```python
tuple1 = ('Foo',34, 5.0, 34)
print(tuple1.count(34))
```

### 3.5 Secuencias
- Formalmente, objetos iterables no materializados
- Noson listas. ‘list()’ para materializarla
- ‘range’ o ‘enumerate'

```python
list(range(0,10,2))
lista = [10, 20, 30]
print(list(enumerate(lista)))
```

```python
# enumerate para iterar una colección (índice y valor)

lista = [10, 20, 30] # Crea una lista

for index, value in enumerate(lista): # Recorre la lista con índice y valor
    print('Index = ' + str(index) + '. Value = ' + str(value)) # Muestra el índice y el valor

for index, value in [(0, 10), (1, 20), (2, 30)]: # Recorre una lista de tuplas
    print('Index = ' + str(index) + '. Value = ' + str(value)) # Muestra el índice y el valor
```

```python
# enumerate para iterar una colección (índice y valor)

for index, value in enumerate(range(0, 10, 2)):
    print('Index = ' + str(index) + '. Value = ' + str(value))

for value in enumerate(range(0, 10, 2)):
    print(value)
```

```python
# zip para unir, elemento a elemento, dos colecciones,
# retornando una lista de tuplas. Es útil para iterar
# dos listas al mismo tiempo.

nombres = ['Manolo', 'Pepe', 'Luis']
edades = [31, 34, 34, 45]

for nombre, edad in zip(nombres, edades):
    print('Nombre: ' + nombre + ', edad: ' + str(edad))
```

```python
nombres = ['Manolo', 'Pepe', 'Luis']
edades = [31, 34, 34]

for i in range(0, len(nombres)):
    print(f"Nombre es {nombres[i]} y edad es {edades[i]}")
```

```python
nombres = ['Manolo','Pepe','Luis']
edades = [31,34,34]
jugadores = zip(nombres,edades) # genera secuencia
print(list(jugadores))
print(type(jugadores))
```

```python
# listas de diferentes longitudes
nombres = ['Manolo','Pepe','Luis']
edades = [31,34,34,44,33]
jugadores = zip(nombres,edades)
print(list(jugadores))
```

```python
# unzip

nombres = ['Manolo', 'Pepe', 'Luis'] # Crea una lista de nombres
edades = [31, 34, 34] # Crea una lista de edades
alturas = [120, 130, 140] # Crea una lista de alturas

jugadores = list(zip(nombres, edades, alturas)) # Une las listas en una lista de tuplas

print(jugadores) # Muestra la lista de tuplas

ns, es, al = zip(*jugadores) # Separa las tuplas en listas independientes

print(list(ns)) # Muestra la lista de nombres
print(list(es)) # Muestra la lista de edades
print(list(al)) # Muestra la lista de alturas
```

```python
nombres = ['Manolo', 'Pepe', 'Luis'] # Crea una lista de nombres
edades = [31, 34, 34] # Crea una lista de edades
dd = [23, 34, 45] # Crea una tercera lista

jugadores_z = zip(nombres, edades, dd) # Une las listas en un iterador de tuplas
print(list(jugadores_z)) # Muestra las tuplas generadas

jugadores_uz = zip(*jugadores_z) # Intenta separar las tuplas
print(list(jugadores_uz)) # No muestra datos porque el iterador ya fue consumido
```

### 3.6 Para terminar, volvemos a enfatizar un matiz importante …
    ### En Python todo son objetos
Cada objeto tiene:
- Identidad: Nunca cambia una vez creado. Es como la dirección de memoria. Operador is
compara identidad. Función id() devuelve identidad.
- Tipo: determina posibles valores y operaciones. Función type() devuelve el tipo. No cambia.
- Valor: que pueden ser mutables e inmutables.
    - Tipos mutables: list, dictionary, set y tipos definidos por el usuario.
    - Tipos inmutables: int, float, bool, string y tupl

```python
# Asignación

list_numbers = [1, 2, 3]      # Lista (mutable)
tuple_numbers = (1, 2, 3)     # Tupla (inmutable)

print(list_numbers[0])
print(tuple_numbers[0])

# Modificar la lista
list_numbers[0] = 100

# Para modificar una tupla:
tuple_l_numbers = list(tuple_numbers)
tuple_l_numbers[0] = 100
tuple_numbers = tuple(tuple_l_numbers)

print(list_numbers)
print(tuple_numbers)
```

```python
# Identidad

list_numbers = [1, 2, 3]
tuple_numbers = (1, 2, 3)

print('Id list_numbers: ' + str(id(list_numbers)))
print('Id tuple_numbers: ' + str(id(tuple_numbers)))

list_numbers += [4, 5, 6] # La lista original se extiende
tuple_numbers += (4, 5, 6) # Se crea un nuevo objeto

print(list_numbers)
print(tuple_numbers)

print('Id list_numbers: ' + str(id(list_numbers)))
print('Id tuple_numbers: ' + str(id(tuple_numbers)))
```

```python
# Referencias

list_numbers = [1, 2, 3]
list_numbers_2 = list_numbers # list_numbers_2 referencia a list_numbers

print('Id list_numbers: ' + str(id(list_numbers)))
print('Id list_numbers_2: ' + str(id(list_numbers_2)))

list_numbers.append(4) # Se actualiza list_numbers2 también

print(list_numbers)
print(list_numbers_2)

print('Id list_numbers: ' + str(id(list_numbers)))
print('Id list_numbers_2: ' + str(id(list_numbers_2)))
```

```python
text = "Hola" # Crea una cadena de texto
text_2 = text # Asigna la referencia de la cadena

print('Id text: ' + str(id(text))) # Muestra la dirección en memoria de text
print('Id text_2: ' + str(id(text_2))) # Muestra la dirección en memoria de text_2

text += " Mundo" # Crea una nueva cadena concatenando texto

print(text) # Muestra la nueva cadena
print(text_2) # Muestra la cadena original

print('Id text: ' + str(id(text))) # Muestra la nueva dirección en memoria de text
print('Id text_2: ' + str(id(text_2))) # Muestra la dirección en memoria de text_2
```

```python
teams = ["Team A", "Team B", "Team C"] # Crea una lista
player = (23, teams) # Crea una tupla con una lista

print(type(player)) # Muestra el tipo de dato
print(player)       # Muestra la tupla
print(id(player))   # Muestra la dirección en memoria

teams[2] = "Team J" # Modifica un elemento de la lista

print(player)      # Muestra la tupla actualizada
print(id(player))  # Muestra la dirección en memoria

player[1][0] = 'Team XX' # Modifica un elemento de la lista dentro de la tupla
print(teams) # Muestra la lista modificada
```

```python
# Ejercicio 5

# Escribe un programa que, dada una lista de strings L, un string s perteneciente a L y un string t,
# reemplace s por t en L. El programa debe mostrar la lista resultante por pantalla.

L = ["Python", "Java", "C++", "Kotlin"] # Crea una lista de lenguajes
s = "Java"                              # Define el elemento a reemplazar
t = "JavaScript"                        # Define el nuevo valor

indice = L.index(s)     # Obtiene el índice del elemento
L[indice] = t           # Reemplaza el elemento

print(L)                # Muestra la lista actualizada
```

```python
# Ejercicio 6

# Escribe un programa que defina una tupla con elementos numéricos, reemplace el valor del último por un valor diferente y muestre
# la tupla por pantalla. Recuerda que las tuplas son inmutables. Tendrás que usar objetos intermedios.

tupla = (1,2,3,4) # Crea una tupla

lista = list(tupla) # Convierte la tupla en lista
lista[-1] = 5 # Modifica el último elemento

print(tuple(lista)) # Convierte la lista en tupla y la muestra
```

```python
# Ejercicio 8

# Escribe un programa que, dada una tupla y un índice válido i, elimine el elemento de la tupla que se encuentra en la posición i.
# Para este ejercicio sólo puedes usar objetos de tipo tupla. No puedes convertir la tupla a una lista, por ejemplo.

tupla = ('a', 'b', 'c', 'd', 'e', 'f') # Crea una tupla
i = 3                                  # Define el índice a eliminar

print(tupla[:i] + tupla[i+1:]) # Crea una nueva tupla sin el elemento indicado
```

```python
# Ejercicio 9

# Escribe un programa que obtenga la mediana de una lista de números. Recuerda que la mediana M de una lista de números L
# es el número que cumple la siguiente propiedad: la mitad de los números de L son superiores a M y la otra mitad son inferiores.
# Cuando el número de elementos de L es
# par, se puede considerar que hay dos medianas. No obstante, en este ejercicio consideraremos que únicamente existe una mediana.

L = [21, 9, 1, 5.5, 19, 3, 15] # Crea una lista de números

L.sort()       # Ordena la lista de menor a mayor
print(L)      # Muestra la lista ordenada

mediana = L[len(L)//2] # Obtiene el elemento central

print(mediana)         # Muestra la mediana
```

### 1.5 Paso de parámetros
- Tipos simples (inmutables) por valor: int, float, string.
  - El efecto es como si se creara una copia dentro de la función, aunque realmente no es
esto lo que ocurre.
  - Los cambios dentro de la función no afectan fuera

```python
def cambiar_valor(x):
    x += 3
    print('Dentro de la función:', x) # x ahora referencia a un nuevo objeto␣ (el 3), pero esto no afecta a la variable de fuera de la función, que sigue␣ apuntando al 2.
    return

a = 2
cambiar_valor(a)
print('Fuera de la función:', a)
```

- Tipos complejos (mutables) por referencia: list, set, dictionary.
  - Dentro de la función se maneja el mismo objeto que se ha pasado desde fuera.
  - Los cambios dentro de la función sí afectan fuera.

```python
def anyadir_2_a(x):
    x.append(4)

lista = [0, 1,3]
anyadir_2_a(lista)
print(lista)
```

```python
# with tuples
def anyadir_2_a(x):
    print(id(x))
    return

lista = (0, 1)
print(id(lista))
anyadir_2_a(lista)
print(lista)
```

- Siquiero que no se modifique un objeto de un tipo complejo, los cuales se pasan por referencia, puedo pasar una copia a la función:

```python
def anyadir_2_a(x):
    x.append(2)
    print('Dentro de la función: ', x)

lista = [0, 1]
anyadir_2_a(lista.copy())
print('Fuera de la función: ', lista)
```

- Si la lista tiene sublistas anidadas hay que hacer un deep copy

```python
# Importar el módulo copy para realizar copias profundas.
import copy

# Lista con elementos simples y una lista anidada.
lista = [2, 4, 16, 32, [34, 10, [5, 5]]]

# Copia superficial (comentada como referencia).
copia = lista.copy()

# Crear una copia profunda de la lista.
# Esto permite modificar la copia sin afectar la lista original.
copia = copy.deepcopy(lista)

# Modificar un elemento simple de la copia.
copia[0] = 454

# Modificar un elemento dentro de una lista anidada.
copia[4][2][0] = 64

# Mostrar la lista original.
print(lista)

# Mostrar la copia modificada.
print(copia)

# Mostrar el identificador (dirección de memoria) de ambas listas.
print(f"{id(lista)} - {id(copia)}")

# Mostrar el identificador de las listas anidadas.
# Al usar deepcopy(), ambas tienen direcciones diferentes.
print(f"{id(lista[4])} - {id(copia[4])}")
```

```python
# Función que construye una consulta SQL utilizando
# una cantidad variable de argumentos con nombre (**kwargs).

def consulta_sql(**kwargs):
    
    # Mostrar el diccionario recibido como parámetro.
    print(kwargs)
    
    # Consulta SQL base.
    sql = "SELECT * FROM bicis WHERE "
    
    # Crear una lista con las condiciones recibidas.
    # Ejemplo: station=alameda, free=10
    args = [f"{key}={value}" for key, value in kwargs.items()]
    
    # Unir las condiciones mediante el operador AND.
    sql += " AND ".join(args)
    
    # Devolver la consulta completa.
    return sql

# Llamada a la función utilizando argumentos con nombre.
print(consulta_sql(station='alameda', free=10))
```

### 1.7 Recursividad
- Funciones que se llaman a sí mismas

```python
# Función recursiva para calcular el factorial de un número.
# El caso base ocurre cuando n es igual a 0.

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Mostrar el factorial de 10 utilizando recursividad.
print(factorial(10))


# Función iterativa para calcular el factorial de un número.
# Utiliza un ciclo "for" para multiplicar los valores desde 1 hasta n.

def fact(n):
    a = 1

    for i in range(1, n + 1):
        a *= i

    return a

# Mostrar el factorial de 10 utilizando un ciclo.
print(fact(10))
```

```python
fact = lambda x: 1 if x == 0 else x * fact(x-1)
fact(10)
```

### 1.8 Funciones Lambda
- Es posible definir funciones anónimas (sin nombre).
- Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una sentencia def
no puede (por ejemplo, dentro de una lista o como parámetro de una función).
- Útiles cuando se quiere pasar una función como argumento a otra. <br>

Formato general: <br>
lambda `<lista de argumentos> : <valor a retornar>`
- Importante: `<valor a retornar>` no es un conjunto de instrucciones. Es simplemente una
expresión return que omite esta palabra.
- Las funciones definidas con def son más generales:
   - Cualquier cosa que implementes en un lambda, lo puedes implementar como una función
convencional con def, pero no viceversa.

```python
# Ejemplo de una función tradicional que recibe tres parámetros y devuelve su suma.

def suma(x, y, z):
    return x + y + z

# Llamada a la función tradicional.
print(suma(1, 2, 3))


# Ejemplo de una función lambda.
# Las funciones lambda permiten definir funciones pequeñas en una sola línea.

suma2 = lambda x, y, z: x + y + z

# Llamada a la función lambda.
print(suma2(1, 2, 3))
```

```python
# La función 'sort()' puede recibir una función opcional mediante el parámetro 'key'.
# Esta función se ejecuta sobre cada elemento de la lista
# antes de realizar las comparaciones para el ordenamiento.

# Lista de ciudades.
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

# Ordenar alfabéticamente (por defecto).
cities.sort()
print(cities)

# Ordenar por la cantidad de caracteres utilizando una función lambda.
cities.sort(key=lambda x: len(x))
print(cities)

# Ordenar por la cantidad de caracteres utilizando la función integrada len.
cities.sort(key=len)
print(cities)

# Definir una función que devuelve la longitud de un texto.
def longitud(x):
    return len(x)

# Ordenar utilizando una función definida por el usuario.
cities.sort(key=longitud)
print(cities)
```

```python
# Ejemplo de ordenamiento sin utilizar funciones lambda.

# Lista de ciudades.
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

# Función que devuelve la cantidad de caracteres
# de una cadena de texto.
def num_caracteres(x):
    return len(x)

# Ordenar la lista utilizando la función definida anteriormente.
cities.sort(key=num_caracteres)

# Mostrar la lista ordenada.
print(cities)
```

- Diccionarios se pueden utilizar como ‘switch’ en otros lenguajes
- Ideal para escoger entre varias opciones

```python
# Función que simula una estructura "switch" utilizando un diccionario y funciones lambda.
def switch_dict(operator, x, y):
    # Diccionario con las operaciones disponibles.
    # Según el valor de "operator", se ejecutará la función correspondiente.
    return {
        'add': lambda: x + y,
        'sub': lambda: x - y,
        'mul': lambda: x * y,
        'div': lambda: x / y,
    }.get(operator, lambda: None)()   # Retorna None si no encuentra la operación.

# Ejemplos de uso.

# Suma
print(switch_dict('add', 2, 2))

# División
print(switch_dict('div', 10, 2))

# Operación inexistente
print(switch_dict('mod', 17, 3))
```

### 1.9 Scopes
- La localización de una asignación a una variable en el código determina desde donde puedes
acceder a esa variable.
- Porejemplo, una variable asignada dentro de una función sólo es visible dentro de esa función.
- El alcance de la visibilidad de una variable determina su scope.
- El término scope hace referencia a un espacio de nombres (namespace). Por ejemplo, una
función establece su propio namespace.

```python
# Variable global.
# Puede ser utilizada desde cualquier parte del programa
# excepto cuando una variable local con el mismo nombre la oculta.

X = 10

# Función que contiene una variable local.
def func():
    X = 20  # Variable local a la función.

    # Función interna.
    def func_int():
        X = 30  # Variable local a la función interna.
        print(X)

    # Llamada a la función interna.
    func_int()

# Llamada a la función principal.
func()

# Mostrar el valor de la variable global.
print(X)
```

### Regla LEGB Dada una función F, tenemos los siguientes scopes:
- Local: variables locales a F.
- Enclosing: variables localizadas en funciones que contienen a (o por encima de) F.
- Global: variables definidas en el módulo (fichero) que no están contenidas en ninguna función. <br>
Este scope no abarca más de un fichero.
- Built-ins: proporcionadas por el lenguaje. <br>

La resolución de nombres ocurre de abajo a arriba.

```python
# Variable global.
# Es accesible desde cualquier parte del programa.

X = 99

# Función que recibe un parámetro local.
def func(Y):
    
    # Variable local.
    # Utiliza la variable global X y el parámetro Y.
    Z = X + Y
    
    # Devuelve el resultado de la suma.
    return Z

# Llamada a la función.
print(func(1))
```

```python
X = 1
def func():
    X = 2 # X es una variable diferente

func()
print(X)
```

- La sentencia global nos permite modificar una variable global desde dentro de una función

```python
# Variable global.
X = 1

# Función que modifica la variable global.
def func():
    
    # Indica que se utilizará la variable global X
    # y no se creará una nueva variable local.
    global X
    
    # Modificar el valor de la variable global.
    X = 2

# Llamada a la función.
func()

# Mostrar el nuevo valor de la variable global.
print(X)
```

- No hay necesidad de usar global para referenciar variables. Sólo es necesario para modificarlas.

```python
# Variables globales.
y, z = 1, 2

# Función que crea una nueva variable global.
def todas_globales():
    global x

    # x utiliza las variables globales y y z.
    x = y + z

# Llamada a la función.
todas_globales()

# Mostrar las variables globales.
print(x)
print(y)
print(z)
```

## Closures
- Las funciones pueden recordar su enclosing scope, independientemente de si éstos continuan
existiendo o no.

```python
def f1():
    X = 88       # Enclosing scope (para f2)
    def f2():
        print(X)
    return f2     # Devuelve f2, sin invocarla

accion = f1()
accion()
```

- Sentencia nonlocal para modificar variables que no son locales, pero tampoco globales

```python
# Ejemplo del uso de la palabra reservada "nonlocal".
# Permite modificar una variable de la función externa
# (Enclosing Scope) desde una función interna.

def f1():

    # Variable del ámbito de la función f1.
    contador = 10

    # Función interna.
    def f2():

        # Indica que se utilizará la variable "contador"
        # definida en la función externa.
        nonlocal contador

        # Incrementar el contador.
        contador += 1

        # Mostrar el valor actualizado.
        print(contador)

    # Retornar la función interna.
    return f2


# Guardar la función retornada.
accion = f1()

# Ejecutar la función varias veces.
accion()
accion()
accion()
```

1. Escribe una función que reciba como entrada una lista con números y devuelva como resultado
una lista con los cuadrados de los números contenidos en la lista de entrada.

```python
# Función que devuelve una lista con los cuadrados
# de los números de la lista de entrada.
def cuadrados(lista):

    # Crear una lista vacía para almacenar los resultados.
    resultado = []

    # Recorrer cada número de la lista.
    for numero in lista:

        # Calcular el cuadrado y agregarlo a la nueva lista.
        resultado.append(numero ** 2)

    # Devolver la lista resultante.
    return resultado


# Lista de ejemplo.
numeros = [2, 4, 6, 8, 10]

# Mostrar el resultado.
print(cuadrados(numeros))
```

2. Escribe una función que reciba números como entrada y devuelva la suma de los mismos. La
función debe ser capaz de recibir una cantidad indeterminada de números. La función no
debe recibir directamente ningún objeto complejo (lista, conjunto, etc.).

```python
# Función que recibe una cantidad variable de números
# y devuelve la suma de todos ellos.
def suma(*numeros):

    # Inicializar el acumulador.
    total = 0

    # Recorrer todos los números recibidos.
    for numero in numeros:
        total += numero

    # Devolver la suma total.
    return total


# Ejemplos de uso.
print(suma(5))
print(suma(2, 4))
print(suma(1, 3, 5, 7))
print(suma(10, 20, 30, 40, 50))
```

3. Escribe una función que reciba un string como entrada y devuelva el string al revés. Ejemplo:
si el string de entrada es ‘hola’, el resultado será ‘aloh’.

```python
# Función que devuelve un string invertido.
def invertir_cadena(texto):

    # Variable donde se almacenará la cadena invertida.
    invertida = ""

    # Recorrer la cadena desde el último carácter hasta el primero.
    for i in range(len(texto) - 1, -1, -1):
        invertida += texto[i]

    # Devolver la cadena invertida.
    return invertida


# Ejemplo de uso.
palabra = "hola"

print(invertir_cadena(palabra))
```

(por ejemplo, un múmero menor o igual a cero), mostrará un mensaje de error por pantalla

```python
# Función que devuelve los cinco primeros múltiplos
# de un número entero positivo.
def multiplos(numero):

    # Verificar que el número sea válido.
    if numero <= 0:
        print("Error: el número debe ser un entero positivo.")
        return []

    # Crear una lista vacía para almacenar los múltiplos.
    lista = []

    # Generar los cinco primeros múltiplos.
    for i in range(1, 6):
        lista.append(numero * i)

    # Devolver la lista de múltiplos.
    return lista


# Ejemplos de uso.
print(multiplos(3))
print(multiplos(7))
print(multiplos(-2))
```

7. Escribe una función que reciba una lista como parámetro y compruebe si la lista tiene duplicados. La función devolverá True si la lista tiene duplicados y False si no los tiene.

```python
# Función que verifica si una lista contiene elementos duplicados.
def tiene_duplicados(lista):

    # Lista auxiliar para guardar los elementos ya vistos.
    elementos = []

    # Recorrer cada elemento de la lista.
    for elemento in lista:

        # Si el elemento ya existe, significa que está duplicado.
        if elemento in elementos:
            return True

        # Si no existe, agregarlo a la lista auxiliar.
        elementos.append(elemento)

    # Si termina el recorrido sin encontrar duplicados.
    return False


# Ejemplos de uso.
print(tiene_duplicados([1, 2, 3, 4]))
print(tiene_duplicados([1, 2, 3, 2]))
print(tiene_duplicados(["Python", "Java", "Python"]))
```

8. Escribe una función lambda que, al igual que la función desarrollada en el ejercicio anterior, reciba una lista como parámetro y compruebe si la lista tiene duplicados. La función devolverá True si la lista tiene duplicados y False si no los tiene.

```python
# Función lambda que verifica si una lista tiene elementos duplicados.
tiene_duplicados = lambda lista: len(lista) != len(set(lista))

# Ejemplos de uso.
print(tiene_duplicados([1, 2, 3, 4]))
print(tiene_duplicados([1, 2, 3, 2]))
print(tiene_duplicados(["Python", "Java", "Python"]))
```

9. Escribe una función que compruebe si un string dado es un palíndromo. Un palíndromo es
una secuencia de caracteres que se lee igual de izquierda a derecha que de derecha a izquierda.
Por ejemplo, la función devolverá True si recibe el string “reconocer” y False si recibe el string
“python”.

```python
# Función que verifica si una cadena es un palíndromo.
def es_palindromo(texto):

    # Comparar la cadena original con su versión invertida.
    if texto == texto[::-1]:
        return True
    else:
        return False


# Ejemplos de uso.
print(es_palindromo("reconocer"))
print(es_palindromo("radar"))
print(es_palindromo("python"))
```

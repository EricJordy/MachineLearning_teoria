## 1 Acceso a ficheros
### 1.1 Ficheros
- File es un tipo de objeto predefinido en Python (built-in).
- Permite acceder a ficheros desde programas Python.
- Los ficheros son de un tipo especial:
   - Son built-in, pero no son ni números, ni secuencias, ni mappings. Tampoco responden a
operadores en expresiones.
- La función open permite crear objetos de tipo fichero. <br>

Formato general para abrir un fichero:<br>

afile = open(filename, mode)
- mode es opcional. Por defecto, los ficheros se abren en modo lectura.
-  Los datos leidos de un fichero siempre se obtienen en formato string. Lo mismo ocurre con
escritura.
- Los ficheros se deben cerrar invocando close (liberación de recursos).

```python
# Abrir el archivo ubicado dentro de la carpeta "res".
# Por defecto, open() abre el archivo en modo lectura ("r").

mi_fichero = open("res/multiple_lines.txt")

# Leer todo el contenido del archivo.
# El método read() devuelve un único string con todo el texto.

print(mi_fichero.read())

# Cerrar el archivo para liberar los recursos del sistema.

mi_fichero.close()
```

```python
# Lectura línea por línea utilizando un bucle for.

# Abrir el archivo en modo lectura.
mi_fichero = open("res/multiple_lines.txt")

# Recorrer el archivo una línea a la vez.
for linea in mi_fichero:
    print(f"{linea}")

# Cerrar el archivo.
mi_fichero.close()
```

```python
# Leer todas las líneas del archivo y almacenarlas en una lista.

# Abrir el archivo en modo lectura.
mi_fichero = open('res/multiple_lines.txt')

# Leer todas las líneas del archivo.
# Cada línea será un elemento de la lista.
lineas = mi_fichero.readlines()

# Cerrar el archivo.
mi_fichero.close()

# Mostrar la lista completa.
print(lineas)
```

- Resolución de paths independiente de plataforma

```python
# Importar el módulo "os".
# Este módulo permite trabajar con archivos, carpetas y rutas del sistema.

import os

# Crear la ruta hacia el archivo "multiple_lines.txt".
# os.path.join() une automáticamente las carpetas utilizando
# el separador correcto según el sistema operativo.

ruta = os.path.join("res", "multiple_lines.txt")

# Mostrar la ruta generada.
print(ruta)
```

```python
# Lista que contiene cada parte de la ruta.
dataset_file_path = ["res", "multiple_lines.txt"]

# El operador * desempaqueta la lista y envía
# cada elemento como un argumento independiente.
ruta = os.path.join(*dataset_file_path)

# Mostrar la ruta creada.
print(ruta)
```

- close autómatico con sentencia with. Esta es la forma habitual de leer de fichero en Python,
context manager

```python
# Abrir el archivo utilizando un Context Manager.
# El archivo se cerrará automáticamente al salir del bloque "with".

with open(ruta) as mi_fichero:

    # Recorrer el archivo línea por línea.
    for linea in mi_fichero:

        # Mostrar cada línea sin agregar un salto de línea adicional.
        print(linea, end="")
```

- Se puede abrir varios ficheros en un mismo with.

```python
# Crear las rutas de los dos archivos.

ruta1 = os.path.join("res", "one_line.txt")
ruta2 = os.path.join("res", "multiple_lines.txt")

# Abrir ambos archivos utilizando un único Context Manager.
with open(ruta1) as fichero1, open(ruta2) as fichero2:

    # Leer todas las líneas del primer archivo.
    print(fichero1.readlines())

    # Leer todas las líneas del segundo archivo.
    print(fichero2.readlines())
```

### 1.3 Acceso para escritura

```python
# Función que crea una lista de números en formato texto.
# Cada número termina con un salto de línea (\n).

def crear_lista(tamanyo):

    lista = []

    # Generar los números desde 0 hasta tamanyo - 1.
    for i in range(tamanyo):
        lista.append(str(i) + "\n")

    return lista


# Crear la ruta donde se guardará el archivo.
ruta = os.path.join("res", "a_dummy.txt")

# Abrir el archivo en modo escritura de texto ("wt").
with open(ruta, "w") as fichero:

    # Escribir la cabecera.
    fichero.write("Cabecera del ciclo del for en el ejemplo\n")

    # Crear la lista de números.
    lista = crear_lista(10)

    # Escribir todas las líneas de la lista.
    fichero.writelines(lista)

# Confirmar que el archivo fue creado.
print("Archivo creado correctamente.")
```

## Buffering
- Por defecto, el texto que transfieres desde tu programa a un fichero no se guarda en disco
inmediatamente. Se almacena en un buffer.
- Acciones como cerrar un fichero o invocar el método flush fuerzan que se transfiera el con
tenido del buffer a disco.

```python
# Crear la ruta del archivo.
ruta = os.path.join("res", "FicheroParaEscritura.txt")

# Abrir el archivo en modo escritura.
fichero_write = open(ruta, "w")

# Escribir un texto en el buffer.
fichero_write.write("foter")

# Abrir el mismo archivo en modo lectura.
fichero_read = open(ruta, "r")

# Intentar leer el contenido antes de vaciar el buffer.
print(fichero_read.readlines())

# Forzar que el contenido del buffer se escriba en el disco.
fichero_write.flush()

# Volver al inicio del archivo para leer nuevamente.
fichero_read.seek(0)

# Leer nuevamente el contenido.
print(fichero_read.readlines())

# Cerrar ambos archivos.
fichero_write.close()
fichero_read.close()
```

### 1.4 Archivos CSV
- Python permite leer datos de ficheros CSV y también escribir ficheros en este formato.
- Popular formato en ciencia de datos

```python
# El archivo tabla_operaciones.csv contiene datos
# separados por comas.

import csv
import os

# Crear la ruta del archivo CSV.
ruta = os.path.join("res", "tabla_operaciones.csv")

# Abrir el archivo utilizando un Context Manager.
with open(ruta) as fichero:

    # Crear un lector de archivos CSV.
    # delimiter="," indica que los campos están separados por comas.
    data_reader = csv.reader(fichero, delimiter=",")

    # Recorrer cada fila del archivo.
    for linea in data_reader:

        # Mostrar las dos primeras columnas.
        print(linea[0] + " ---- " + linea[1])
```

```python
import csv
import os

# Ruta del archivo CSV original.
ruta = os.path.join("res", "tabla_operaciones.csv")

# Abrir el archivo para lectura.
f = open(ruta)

# Crear el lector CSV utilizando la coma como separador.
data_reader = csv.reader(f, delimiter=",")

# Ruta del nuevo archivo que se creará.
ruta_o = os.path.join("res", "4_tabla_operaciones.csv")

# Abrir el nuevo archivo en modo escritura.
with open(ruta_o, "w", newline="") as f2:

    # Crear un escritor CSV utilizando "|" como separador.
    csv_writer = csv.writer(f2, delimiter="|")

    # Recorrer todas las filas del archivo original.
    for line in data_reader:

        # Mostrar la fila actual.
        print(line)

        # Escribir la fila en el nuevo archivo.
        csv_writer.writerow(line)

# Cerrar el archivo de lectura.
f.close()

print("Archivo CSV creado correctamente.")
```

## 2 Docstrings
- Python permite adjuntar documentación a los objetos e inspeccionarla a través de la línea de
comandos o durante la ejecución del programa.
- Los docstrings se almacenan en el atributo `__doc__` de cada objeto.
- El valor de dicho atributo se puede mostrar por medio de la función help

```python
help(open)
```

- Noes necesario editar este atributo directamente.
- Para asociar un docstring a un objeto basta con escribir el texto (entre triples comillas) al
principio de los modulos, funciones o clases, antes del codigo ejecutable

```python
def funcion_de_prueba():
    """Esta es la documentación de la función de prueba"""
    pass

help(funcion_de_prueba)
```

- Accediendo al atributo `__doc__` sólo se obtiene el docstring.

```python
print(funcion_de_prueba.__doc__)
```

Como se puede observar, help añade información adicional. Por ejemplo, para entidades más grandes, help muestra el docstring dividido en secciones

```python
import sys
help(sys)
```

Los docstrings que ocupan más de una línea suelen tener estas partes:
- Resumen en la primera línea.
- Línea en blanco.
- Descripción detallada.
- Otra línea en blanco antes del código

```python
# Función de ejemplo con un docstring de varias líneas.

def funcion_de_prueba2():
    """
    Esta es la línea para el resumen.

    Este es el párrafo donde se puede escribir una descripción más
    detallada de la función. Observa que el resumen y la descripción
    detallada van separados por una línea en blanco. También hay
    otra línea en blanco antes de empezar el código.
    """

    # La función no realiza ninguna acción.
    pass

# Mostrar la ayuda de la función.
help(funcion_de_prueba2)
```

1. Escribe una función que reciba una ruta de un fichero de texto y una cadena de caracteres a
buscar y determine si la cadena aparece en el fichero

```python
import os

# Función que busca una cadena dentro de un archivo de texto.
def buscar_cadena(ruta, cadena):

    # Abrir el archivo en modo lectura.
    with open(ruta, "r") as fichero:

        # Leer todo el contenido del archivo.
        contenido = fichero.read()

        # Verificar si la cadena existe dentro del texto.
        if cadena in contenido:
            return True
        else:
            return False


# Crear la ruta del archivo.
ruta = os.path.join("res", "multiple_lines.txt")

# Cadena que se desea buscar.
texto_buscar = "lineas"

# Mostrar el resultado.
if buscar_cadena(ruta, texto_buscar):
    print(f"La cadena '{texto_buscar}' fue encontrada en el archivo.")
else:
    print(f"La cadena '{texto_buscar}' NO fue encontrada en el archivo.")
```

2. Escribe una función que reciba una lista, una ruta destino y un número n. La función debe crear un fichero en la ruta especificada. El contenido del fichero serán los primeros n elementos de la lista. La función debe controlar de manera apropiada los posibles valores de n que estén
fuera de rango.

```python
import os

# Función que guarda los primeros n elementos de una lista
# en un archivo de texto.
def guardar_lista(lista, ruta, n):

    # Verificar que n sea válido.
    if n < 0 or n > len(lista):
        print("Error: el valor de n está fuera del rango de la lista.")
        return

    # Crear el archivo y escribir los primeros n elementos.
    with open(ruta, "w") as fichero:

        for i in range(n):
            fichero.write(str(lista[i]) + "\n")

    print("Archivo creado correctamente.")


# Lista de ejemplo.
lenguajes = ["Python", "Java", "C++", "JavaScript", "C#"]

# Ruta donde se guardará el archivo.
ruta = os.path.join("res", "lenguajes.txt")

# Número de elementos que se desean guardar.
n = 3

# Llamar a la función.
guardar_lista(lenguajes, ruta, n)
```

3. Escribe una función que reciba una ruta de un fichero de texto devuelva un diccionario con la
frecuencia de aparición de cada palabra. Ejemplo: un fichero que contenga la frase ‘es mejor
que venga que que no venga’ devolverá el siguiente diccionario: {‘es’ : 1, ‘mejor’ : 1, ‘que’ :
3, ‘venga’ : 2, ‘no’ : 1}. Para dividir un string en palabras puedes hacer uso del método split

```python
import os

# Función que cuenta la frecuencia de cada palabra
# dentro de un archivo de texto.
def contar_palabras(ruta):

    # Diccionario donde se almacenará el resultado.
    frecuencias = {}

    # Abrir el archivo en modo lectura.
    with open(ruta, "r") as fichero:

        # Leer todo el contenido del archivo.
        contenido = fichero.read()

        # Dividir el texto en palabras.
        palabras = contenido.split()

        # Recorrer cada palabra.
        for palabra in palabras:

            # Si la palabra ya existe, incrementar el contador.
            if palabra in frecuencias:
                frecuencias[palabra] += 1

            # Si aparece por primera vez, agregarla al diccionario.
            else:
                frecuencias[palabra] = 1

    # Devolver el diccionario.
    return frecuencias


# Ruta del archivo.
ruta = os.path.join("res", "texto.txt")

# Mostrar el resultado.
print(contar_palabras(ruta))
```
